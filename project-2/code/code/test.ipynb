```
def read_lines(fname):
  lines = []
  with open(fname, "r") as f:
    for line in f:
      lines.append(line)
  return lines
```

```
def get_unique(data, w="who"):

  unique = set()

  if w == "who":
    search = "pers"
  elif w == "what":
    search = "doin"
  elif w == "where":
    search = "loca"

  for line in data:
    split = line.split()
    for token in split:
      if token[:4] == search:
        unique.add(token)

  return unique

def get_task_length(data):

  task_lengths = []
  for line in data:
    split = line.split()
    task_len = split.count("who")+split.count("what")+split.count("where")
    task_lengths.append(task_len)
  return task_lengths

def get_seq_length(data):

  seq_lengths = []
  for line in data:
    split = line.split()
    seq_lengths.append(len(split))
  return seq_lengths

def get_tasks(data):

  tasks = []
  error = 0
  for line in data:
    split = line.split()
    task_len = split.count("who")+split.count("what")+split.count("where")
    task = ""
    for i in range(task_len*2):
      if split[i] == "who" or split[i] == "what" or split[i] == "where":
        task += (split[i] + " ")
      elif split[i][:4] == "pers":
        task += "person "
      elif split[i][:4] == "loca":
        task += "location "
      elif split[i][:4] == "doin":
        task += "doing "
      else:
        error += 1
    tasks.append(task)

  print(f"{error} errors")
  return tasks
```

```
train_src = read_lines("Data/Copy of train.src")
train_tgt = read_lines("Data/Copy of train.tgt")

test_src = read_lines("Data/Copy of test.src")
test_tgt = read_lines("Data/Copy of test.tgt")
```
