```
def write_data(data, fname):
  with open(fname, "w") as f:
    for line in data:
      f.write(line+"\n")

def read_file(fname):
  lines = []
  with open(fname, "r") as f:
    for line in f:
      lines.append(line[:-1])
  return np.array(lines)

def makedir(run):
  # builds file structure to write results
  base = f"results/random/run_{run}"
  if not os.path.exists("results"):
    os.mkdir("results")
  if not os.path.exists("results/random"):
    os.mkdir("results/random")
  if not os.path.exists(base):
    os.mkdir(base)
    os.mkdir(base + "/All")
    os.mkdir(base + "/Systematicity")
    os.mkdir(base + "/Systematicity/Test1")
    os.mkdir(base + "/Productivity")
    os.mkdir(base + "/Productivity/Tasks")
    os.mkdir(base + "/Productivity/Tasks/Test1")
    os.mkdir(base + "/Productivity/Tasks/Test2")
    os.mkdir(base + "/Productivity/Whens")
    os.mkdir(base + "/Productivity/Whens/Test3")
    os.mkdir(base + "/Productivity/Whens/Test4")
    os.mkdir(base + "/Productivity/Whens/Test5")
```

```
def remove_tasks(split):
  # splits samples into questions and sentence
  www = ["who", "what", "where"]
  i = 0
  tasks = []
  while split[i] in www:
    tasks.append([split[i], split[i+1]])
    i += 2
  return np.array(split[i:]), np.array(tasks)

def get_w(sentence, w):
  # filters words according to w (w = who -> return each person in sentence)
  ws = []
  for t in sentence:
    if (w == "who" and t[:2] == "pe") or (w == "what" and t[:2] == "do") or \
    (w == "where" and t[:2] == "lo"):
      ws.append(t)
  return ws

def translate_random(data):
  # translates sentence according to correct question answer formats (who -> person)
  # but samples the answers randomly from the words in the sentence
  translations = []
  for seq in data:
    sentence, tasks = remove_tasks(seq.split())
    translation = []
    for q, w2 in tasks:
      w1 = np.random.choice(get_w(sentence, q))
      if q == "who" and w2[:2] == "do":
        answer = w1 + " went " + w2
      elif q == "who" and w2[:2] == "lo":
        answer = w1 + " at " + w2
      elif q == "what" and w2[:2] == "pe":
        answer = w2 + " went " + w1
      elif q == "what" and w2[:2] == "lo":
        answer = w1 + " at " + w2
      elif q == "where" and w2[:2] == "pe":
        answer = w2 + " at " + w1
      elif q == "where" and w2[:2] == "do":
        answer = "at " + w1 + " " + w2
      translation.append(answer)
    translations.append(" ".join(translation))
  return translations
```

```
# read all source files
main_src = read_file("data/Final2/All/test5k_all.src")

syst_inc012_src = read_file("data/Final2/Systematicity/Test1/syst1_test_inc012.src")
syst_exc012_src = read_file("data/Final2/Systematicity/Test1/syst1_test_exc012.src")

prod_0tsk_src = read_file("data/Final2/Productivity/Tasks/prod_test_1tasks.src")
prod_1tsk_src = read_file("data/Final2/Productivity/Tasks/prod_test_2tasks.src")
prod_2tsk_src = read_file("data/Final2/Productivity/Tasks/prod_test_3tasks.src")

prod_0whn_src = read_file("data/Final2/Productivity/Whens/prod_test_0whens.src")
prod_1whn_src = read_file("data/Final2/Productivity/Whens/prod_test_1whens.src")
prod_2whn_src = read_file("data/Final2/Productivity/Whens/prod_test_2whens.src")
prod_3whn_src = read_file("data/Final2/Productivity/Whens/prod_test_3whens.src")
prod_4whn_src = read_file("data/Final2/Productivity/Whens/prod_test_4whens.src")
```

```
# for three separate runs, translate all source files and save predictions
runs = [0,1,2]

for run in runs:
  print("run", run)
  makedir(run)
  
  main_prd = translate_random(main_src)

  syst_inc012_prd = translate_random(syst_inc012_src)
  syst_exc012_prd = translate_random(syst_exc012_src)

  prod_0tsk_prd = translate_random(prod_0tsk_src)
  prod_1tsk_prd = translate_random(prod_1tsk_src)
  prod_2tsk_prd = translate_random(prod_2tsk_src)

  prod_0whn_prd = translate_random(prod_0whn_src)
  prod_1whn_prd = translate_random(prod_1whn_src)
  prod_2whn_prd = translate_random(prod_2whn_src)
  prod_3whn_prd = translate_random(prod_3whn_src)
  prod_4whn_prd = translate_random(prod_4whn_src)

  write_data(main_prd, f"results/random/run_{run}/All/test5k_all.prd")

  write_data(syst_inc012_prd, f"results/random/run_{run}/Systematicity/Test1/syst1_test_inc012.prd")
  write_data(syst_exc012_prd, f"results/random/run_{run}/Systematicity/Test1/syst1_test_exc012.prd")
      
  for test in [1,2]:
    write_data(prod_0tsk_prd, f"results/random/run_{run}/Productivity/Tasks/Test{test}/prod_test_1tasks.prd")
    write_data(prod_1tsk_prd, f"results/random/run_{run}/Productivity/Tasks/Test{test}/prod_test_2tasks.prd")
    write_data(prod_2tsk_prd, f"results/random/run_{run}/Productivity/Tasks/Test{test}/prod_test_3tasks.prd")
      
  for test in [3,4,5]:
    write_data(prod_0whn_prd, f"results/random/run_{run}/Productivity/Whens/Test{test}/prod_test_0whens.prd")
    write_data(prod_1whn_prd, f"results/random/run_{run}/Productivity/Whens/Test{test}/prod_test_1whens.prd")
    write_data(prod_2whn_prd, f"results/random/run_{run}/Productivity/Whens/Test{test}/prod_test_2whens.prd")
    write_data(prod_3whn_prd, f"results/random/run_{run}/Productivity/Whens/Test{test}/prod_test_3whens.prd")
    write_data(prod_4whn_prd, f"results/random/run_{run}/Productivity/Whens/Test{test}/prod_test_4whens.prd")
```
